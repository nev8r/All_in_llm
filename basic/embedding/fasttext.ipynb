### fastText

fasttext = subword(n-gram) + word2vec
解决了word2vec 的OOV问题，但是模型复杂度增大

```python
from gensim.models.fasttext import FastText


sentences = [["i", "love", "nlp"], ["fasttext", "is", "great"]]

# 训练 FastText 模型
model = FastText(sentences, vector_size=50, window=3, min_count=1, sg=1, min_n=3, max_n=6)

# 获取词向量
print(model.wv["fasttext"])

# 查看 OOV 词
print(model.wv["fasttexts"])  
```
